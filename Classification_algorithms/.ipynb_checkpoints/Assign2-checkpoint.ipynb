```python
def partition(X, T, ratio=[0.8, 0.2]): 
    
    assert(np.sum(ratio) == 1)
    
    # shuffle the data indices 
    idxs = np.random.permutation(X.index)
    
    # the number of samples 
    N = X.shape[0]
    
    Xs = []
    Ts = []
    i = 0  # first index to zero
    for k, r in enumerate(ratio):
        nrows = int(round(N * r))  # number of rows
        
        if k == len(ratio) -1:
            Xs.append(X.iloc[i:, :])
            Ts.append(T.iloc[i:, :])
        else:
            Xs.append(X.iloc[i:i+nrows, :])
            Ts.append(T.iloc[i:i+nrows, :])
        
        i += nrows
    
    return Xs, Ts

Xlst, Tlst = partition(pd.DataFrame(X_all), pd.DataFrame(T))

Xlstnor, Tlstnor = partition(pd.DataFrame(Xnorx), pd.DataFrame(T))

#pocket:

Xnortrain=Xlstnor[0].head(8000).values
Xnortest=Xlstnor[1].head(2000).values
Tnortrain=Tlstnor[0].head(8000).values
Tnortest=Tlstnor[1].head(2000).values

#full training list
Xtrainin=Xlst[0].values
#training first 4000
###
Xtrainin1=Xlst[0].head(4000).values
print(Xtrainin1.shape)

#full testing list
Xtestin=Xlst[1]
####
Xtestin1=Xlst[1].head(1000).values
print(Xtestin1.shape)


Ttrainin=Tlst[0].head(4000).values
####.head(1000).values
Ttrainin1=Tlst[0]
print(Ttrainin1.shape)
Ttestin=Tlst[1]
####
Ttestin1=Tlst[1].head(1000).values
print(Ttestin1.shape)
```

# III. Method

Review of Models:
---------------------
 
Classification models predict the outcome of the observed values which are usually of the form of class labels. For example, based on the given input a classification model could predict the digit of a handwritten image. It does so by building a model and classifying the image into one of the labels between 0-9.
Some of the algorithms for classification models are:

Pocket Algorithm:
-------------------
The pocket algorithm is an extension of the perceptron algorithm, in which the pocket algorithm stores the best value of w to predict Y. The perceptron algorithm uses the learning rate alpha and the target value to adjust the value of w. But the perceptron algorithm does not guarantee to converge if the data is not linearly separable and the selected decision boundary may not be the best. To overcome the above limitation of perceptron algorithm, pocket algorithm is usually used.


Discriminant Analysis:
----------------------------
Discriminant analysis models the distribution of the predictors X separately in each of the response classes (i.e. default = Yes, default = No), and then uses Bayes theorem to convert these around into estimates for the probability of the response category given the value of X.

QDA:
-----
Quadratic Discriminant Analysis uses different Covariance matrix for each class to calculate the discriminant function  k(x). The decision boundary is placed where the discriminant functions meet such as  1==2.Since the  k function is quadratic in  x , We call this approach as Quadratic Discriminant Analysis (QDA).

LDA:
-----
Linear Discriminant Analysis uses same Covariance matrix for all classes to calculate the discriminant function k(x). Since maintaining a covariance matrix is not cheap and class boundary has higher chance of overfitting in QDA , linear discriminant function is used to overcome this limitation.

Logistic Regression:
------------------------
In logistic regression we predict the probability of the class label P(T=kx) and not the class label k itself. Similar to linear regression we use a gradient decent with learning rate alpha and update the weights based on the predicted output. In logistic regression we use SoftMax function to make sure predicted probabilities add up to 1 and also there are no negative values probabilities.


Implementation of the Codes:
--------------------------------

The data was partitioned into ratio 80:20 for training and testing data set.

The classifier class has an abstract class train and use which is used by all models. The classifier class also provides a normalize and add_ones function for models to normalize data and to add bias.

Each of the models implements the train and use function in their respective child class.

The pocket class train method takes the data and calculates the w value. This w values is used by the use method to calculate the predicted output y.

The QDA class train method takes in the Xp and Xn which are the training data set of different classes and computes two different values of sigma , mu , prior. It also implements a QDA method which calculates the two discriminant function k(x). In the use method ,we compare the k(x) values to classify the output into respective class labels.

The LDA class implementation is similar to the QDA implementation stated above except for the fact the LDA will use the same sigma covariance matrix to calculate the the two discriminant function k(x).The LDA class has a separate LDA method to calculate the discriminant function k(x). In the use method ,we compare the k(x) values to classify the output into respective class labels.

The Logistic class apart from the train and use method also implements the SoftMax function which is gives the probability distribution of the class labels. The train method calculates the value of w using gradient decent method with learning rate alpha and the probability distribution of the class labels. The use method uses the w with test data to give the predicted output Y.



 
 
 
 


### A. Super Classs Definition

```python
import numpy as np
from abc import ABC, abstractmethod

# Super class 

class BaseModel(ABC):
    """ Super class"""
    
    @abstractmethod
    def train(self, X, T):
        pass

    @abstractmethod
    def use(self, X):
        pass

    

class Classifier(BaseModel):
    """
        Abstract class for classification 
        
        Attributes
        ==========
        meanX       ndarray
                    mean of inputs (from standardization)
        stdX        ndarray
                    standard deviation of inputs (standardization)
    """

    def __init__(self, ):
        self.meanX = None
        self.stdX = None

    def normalize(self, X):
        """ standardize the input X """
        
        if not isinstance(X, np.ndarray):
            X = np.asanyarray(X)

        self.meanX = np.mean(X, 0)
        self.stdX = np.std(X, 0)

        # TODO: Finish this normalization
        Xs = (X - self.meanX) / self.stdX
        return Xs

    def _check_matrix(self, mat, name):
        if len(mat.shape) != 2:
            raise ValueError(''.join(["Wrong matrix ", name]))
        
    # add a basis
    def add_ones(self, X):
        """
            add a column basis to X input matrix
        """
        self._check_matrix(X, 'X')
        return np.hstack((np.ones((X.shape[0], 1)), X))

    ####################################################
    #### abstract funcitons ############################
    @abstractmethod
    def train(self, X, T):  
        pass
    
    @abstractmethod
    def use(self, X):
        pass 
```

### B. Pocket Algorithm

```python
from copy import deepcopy as copy 
# Linear Regression Class for least squares
class Pocket(Classifier): 
   
    """ 
        LinearRegress class 
        
        attributes
        ===========
        w    nd.array  (column vector/matrix)
             weights
    """
    def __init__(self):
        Classifier.__init__(self)
        self.w = None
        self.pocket=None
        
    # train lease-squares model
    def train(self, Xs, T):
        T = np.asanyarray(T)
        X1=self.add_ones(Xs)
        X1 = np.asanyarray(X1)
        N = X1.shape[0]
        self.w = np.zeros(X1.shape[1])
        self.w_pocket = copy(self.w)
        maxiter = 100
        alpha = 0.1
            # return 1 if w is better -1 if wp is better
        def compare(X, T, w, wp):
            y = np.sign(X @ w)
            yp = np.sign(X @ wp)
            return 1 if np.sum(y == T) >= np.sum(yp == T) else -1
    
        ###pocket algorithm here

        for i in range(maxiter):
            converged = True
            for k in range(N):
                if(np.dot(X1[k], self.w)*T[k]) <= 0:
                    self.w =self.w + alpha*X1[k]*T[k]
                    converged=False

            if converged:
                print("converged at ", i)
                break
            else:
                best=compare(X1, T, self.w,self.w_pocket)
                if best == 1:
                    self.w_pocket = copy(self.w)
                    
        print("End of training: ", i)
         #plt.plot(X@w_pocket)
    
    
    # apply the learned model to data X
    def use(self, Xuse):
        Xu=self.add_ones(Xuse)
        Y=Xu@self.w_pocket
        return Y

import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline


#Xnortrain=Xlstnor[0].head(8000).values
#Xnortest=Xlstnor[1].head(2000).values
#Tnortrain=Tlstnor[0].head(8000).values
#Tnortest=Tlstnor[1].head(2000).values


#Xtrain=np.vstack((Xtraining_n, Xtraining_p))
#Train=np.vstack((Train_n, Ttrain_p))
pkt=Pocket()
pkt.train(Xnortrain,Tnortrain) 
Y=pkt.use(Xnortest)
Ytrain=pkt.use(Xnortrain)
plt.plot(np.array(Tnortest),'xr')
plt.plot(np.array(Y),'xb')
```

### C. QDA


```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from mpl_toolkits.mplot3d import Axes3D
from copy import deepcopy as copy 


class QDA_class(Classifier): 
   
    def __init__(self):
        Classifier.__init__(self)
        self.Sigma1=None
        self.Sigma2=None
        self.mu1=None
        self.mu2=None
        self.prior1=None
        self.prior2=None
        self.mean=None
        self.stds=None
        
    def train(self,Xn,Xp):
        N1=Xn.shape[0]
        N2=Xp.shape[0]
        N=N1+N2
        Xtrain = np.vstack((Xn, Xp))
        self.means, self.stds = np.mean(Xtrain, 0), np.std(Xtrain, 0)
        Xs=self.normalize(Xtrain)
        self.mu1 = np.mean(Xs[:N1], 0)
        self.mu2 = np.mean(Xs[N1:], 0)
        self.Sigma1 = np.cov(Xs[:N1].T)
        self.Sigma2 = np.cov(Xs[N1:].T)
        self.prior1 = N1 / N
        self.prior2 = N2 / N
        
        
    def QDA(self,X, mu, sigma, prior):
        sigma_inv = np.linalg.inv(sigma)
        diffv = X - mu
        return -0.5  * np.log(np.linalg.det(sigma)) - 0.5 * np.sum(diffv @ sigma_inv * diffv, axis=1) + np.log(prior)
    

    
    
    # apply the learned model to data X
    def use(self, Xuse):
        Xuse1=self.normalize(Xuse)
        d1 = self.QDA(Xuse1, self.mu1, self.Sigma1, self.prior1)
        d2 = self.QDA(Xuse1, self.mu2, self.Sigma2, self.prior2)
        y=np.sign(d1-d2).astype(int)
        return y


    












```

### D. LDA

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from mpl_toolkits.mplot3d import Axes3D
from copy import deepcopy as copy 


class LDA_class(Classifier): 
   
    def __init__(self):
        Classifier.__init__(self)
        self.sigma=None
        self.mu1=None
        self.mu2=None
        self.prior1=None
        self.prior2=None
        self.mean=None
        self.stds=None
        
    def train(self,Xn,Xp):
        N1=Xn.shape[0]
        N2=Xp.shape[0]
        N=N1+N2
        Xtrain = np.vstack((Xn, Xp))
        Xs=self.normalize(Xtrain)
        self.mu1 = np.mean(Xs[:N1], 0)
        self.mu2 = np.mean(Xs[N1:], 0)
        self.sigma = np.cov(Xs.T)
        self.prior1 = N1 / N
        self.prior2 = N2 / N
        
        
    def LDA1(self,X, mu, sigma, prior):
        sigma_inv = np.linalg.inv(sigma)
        return np.sum(np.dot(X, sigma_inv) * mu   - 0.5 * np.dot(mu, sigma_inv) * mu  + np.log(prior), axis=1)

    
    
    # apply the learned model to data X
    def use(self, Xuse):
        Xuse1=self.normalize(Xuse)
        d1 = self.LDA1(Xuse1, self.mu1, self.sigma, self.prior1)
        d2 = self.LDA1(Xuse1, self.mu2, self.sigma, self.prior2)
        y=np.sign(d1-d2).astype(int)
        return y


    












```

```python
def confusion_matrix(Y, T):
    """
        Y    ndarray
             predicted labels
        T    ndarray
             target labels
             
        @cfm DataFrame
             confusion matrix
    """
    
    if len(Y) != len(T):
        raise ValueError("Wrong prediction and target length!")
    
    classes = np.unique(T)
    n_classes = len(classes)
    
    cfm = pd.DataFrame(np.zeros((n_classes, n_classes)), index=classes, columns=classes, dtype=int)
    
    Tidx = [T == c for c in classes]
    for c in classes:
        pred_idx = Y == c
        cfm.loc[c, :] = [np.sum(np.logical_and(pred_idx, tidx)) for tidx in Tidx]
    
    return cfm


print(confusion_matrix(lst1.reshape(test.shape), test))
print("----------------------------------------------")
TP=515
TN=3157
FN=4828
FP=543
Accuracy=((TP+TN)/(TP+FP+TN+FN))*100
import math
print("----------------------------------------------")
print("Accuracy: ",Accuracy , "%")
print("----------------------------------------------")

print("----------------------------------------------")
Precision = (TP/(TP+FP))
print("Precision: ",Precision)
recall=(TP/(TP+FN))
print("Recall: ",recall)
F1=(2*(Precision*recall))/(Precision+recall)
print("F1 score",F1)
Specificity=(TN/(TN+FP))
print("Specificity: ",Specificity)
MCC=((TP*TN)-(FP*FN))/math.sqrt((TP+FP)*(TP+FN)*(TN+FP)*(TN+FN))
print("MCC: ",MCC)
```

### E. Logistic Regression



```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from mpl_toolkits.mplot3d import Axes3D
from copy import deepcopy as copy 


class Logistic_class(Classifier): 
   
    def __init__(self):
        Classifier.__init__(self)
        self.mean=None
        self.stds=None
        self.w=None
        
    def softmax(self,z):
        if not isinstance(z, np.ndarray):
            z = np.asarray(z)
        f = np.exp(z) 
        return f / (np.sum(f, axis=1, keepdims=True) if len(z.shape) == 2 else np.sum(f))

    # for linear fx
    def g(self,X, w):
        return self.softmax(X @ w)          
        
    def train(self,Xn,Xp):
        K=2
        D=Xn.shape[1]
        N1=Xn.shape[0]
        Xtrain = np.vstack((Xn, Xp))
        self.means, self.stds = np.mean(Xtrain, 0), np.std(Xtrain, 0)
        Xs = (Xtrain - self.means) / self.stds
        #Xs=self.normalize(Xtrain)
        N=Xtrain.shape[0]
        Ttrain = np.zeros((N, 2))
        Ttrain[:N1, 0] = 1
        Ttrain[N1:, 1] = 1
        self.w = np.random.rand(D+1, K)
        niter = 1000
        alpha = 0.001
        X1=self.add_ones(Xs)
        for step in range(niter):
            # TODO: add training code here! 
            ys = self.g(X1, self.w)
            self.w += alpha * X1.T @ (Ttrain - ys)
    
    
    # apply the learned model to data X
    def use(self, Xuse):
        Xtests = (Xuse - self.means) / self.stds
        #Xtests=self.normalize(Xuse)      
        X1t = self.add_ones(Xtests)
        Y = self.g(X1t, self.w)
        return Y



    


```

Recall: Is the fraction correct results retrieved compared all positive results available 
F1 Score : which is the metric combing Precision and Recall and closer the F1 score is to 1 the better the result
